```python
# Function
def convert_payment(period):
    #S1
    cond1 = (df[f'min_amount_{period}'] >= df[f'max_amount_{period}'])
    df.loc[cond1,[f'min_amount_{period}',f'max_amount_{period}']] = df.loc[cond1, [f'max_amount_{period}',f'min_amount_{period}']].values
    df[f'min_amount_{period}'].fillna(0.0, inplace = True)
    df[f'max_amount_{period}'].fillna(0.0, inplace = True)
    
    #S2
    cond2 = (df[f'max_amount_{period}'] == 0.0) & df[f'amount_{period}'].isnull()
    df.loc[cond2, [f'amount_{period}']] = 0.0
    
    #S3
    cond3 = (df[f'min_amount_{period}'] == 0) & (df[f'max_amount_{period}'] == 0.0) & df[f'distinct_payment_code_{period}'].isnull()
    df.loc[cond3,[f'distinct_payment_code_{period}']] = 0
    
    #S4
    cond4 = (df[f'distinct_payment_code_{period}'] == 0) & df[f'count_payment_code_{period}'].isnull()
    df.loc[cond4,[f'count_payment_code_{period}']] = 0
    
    #S5
    cond5 = (df[f'distinct_payment_code_{period}'] == 0) & (df[f'count_payment_code_{period}'] == 0) & df[f'distinct_trans_group_{period}'].isnull()
    df.loc[cond5,[f'distinct_trans_group_{period}']] = 0
    
    #S6
    cond6 = (df[f'distinct_payment_code_{period}'] == 0) & (df[f'count_payment_code_{period}'] == 0) & df[f'distinct_ref_no_{period}'].isnull()
    df.loc[cond6,[f'distinct_ref_no_{period}']] = 0
```

## 3.2. Hyperparameter Tuning

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=4)
# StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
best_models = []
val_scores = []
```

### LogisticRegression

```python
# Define the parameter grid for Logistic Regression
param_logit = {
    'penalty' : ['l1', 'l2', 'elasticnet', None],
    'C': np.arange(0.01, 10, 0.1),
    'solver': ['saga','liblinear'],
    'max_iter': [100, 200, 300]
}

# Initialize RandomizedSearchCV with recall as the scoring metric
cv = StratifiedKFold(n_splits=3)
random_search_logit = RandomizedSearchCV(logit, param_distributions=param_logit, n_iter=10, cv=cv, verbose=1, scoring = recall_score, n_jobs=1, random_state=4)

# Fit the RandomizedSearchCV
random_search_logit.fit(X_train_scaled, y_train) # tìm ra bộ siêu tham số hyperparameter tốt nhấ cho algo thuật toán đang chạy trên tập training

best_param_logit = random_search_logit.best_params_ # lấy ra mô hình tốt nhất tìm được ở RandomSearch
print(f"Best parameters found: {best_param_logit}")
```

```python
best_model_logit = random_search_logit.best_estimator_
best_models.append(best_model_logit) #Append in best_models

# Predict with the best model
y_val_logit_pred = best_model_logit.predict(X_val_scaled)
y_val_logit_proba = best_model_logit.predict_proba(X_val_scaled)[:, 1]

# Evaluation
val_recall_logit = recall_score(y_val, y_val_logit_pred) # tính metric của mô hình tốt nhất trên tập validation
val_scores.append(val_recall_logit) # Append in val_scores

val_auc_logit = roc_auc_score(y_val, y_val_logit_proba)
val_mcc_logit = matthews_corrcoef(y_val, y_val_logit_pred)

print(f"Validation Recall Score: {val_recall_logit}")
print(f'Matthews Correlation Coefficient: {val_mcc_logit}')
print(f'AUC Score: {val_auc_logit}')
```

### RandomForestClassifier

```python
# Initialize RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state=4)

# Define the parameter grid for RandomForestClassifier
param_rf = {
    'n_estimators': np.arange(50, 200, 10),
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Initialize RandomizedSearchCV with recall as the scoring metric
cv = StratifiedKFold(n_splits=3)
random_search_rf = RandomizedSearchCV(rf_classifier, param_distributions=param_rf, n_iter=10, cv=cv, verbose=1, scoring=recall_score, n_jobs=-1, random_state=4)

# Fit the RandomizedSearchCV
random_search_rf.fit(X_train_scaled, y_train)

# Best parameters found
best_param_rf = random_search_rf.best_params_
print(f"Best parameters found: {best_param_rf}")
```

```python
best_model_rf = random_search_rf.best_estimator_
best_models.append(best_model_rf) #Append in best_models

# Predict with the best model
y_val_rf_pred = best_model_rf.predict(X_val_scaled)
y_val_rf_proba = best_model_rf.predict_proba(X_val_scaled)[:, 1]

# Validation metrics
val_recall_rf = recall_score(y_val, y_val_rf_pred)
val_scores.append(val_recall_rf) #Append in val_scores

val_auc_rf = roc_auc_score(y_val, y_val_rf_proba)
val_mcc_rf = matthews_corrcoef(y_val, y_val_rf_pred)

print(f"Validation Recall Score: {val_recall_rf}")
print(f'Matthews Correlation Coefficient: {val_mcc_rf}')
print(f'AUC Score: {val_auc_rf}')
```
